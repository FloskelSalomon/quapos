## Filter 95 percentile

- remove everything bigger than 95 percentile of the volume

```python
import numpy as np
import pandas as pd
```

```python
path = "../measurements/wt-postnatal-development/"
```

```python
measurements = pd.read_csv(path + "05-a-measurements-inf-reduced.csv")
measurements
```

```python
filtered_measurements = pd.read_csv(path + "05-b-measurements-filtered-5-percentile.csv")
filtered_measurements
```

```python
print("Number of OS in dataset before filter: ", len(filtered_measurements))
```

```python
ninety_five_percentile = measurements["volume_rescaled"].quantile(0.95)
ninety_five_percentile
```

```python
print("Number of OS in dataset bigger than 95 percentile: ", len(filtered_measurements[filtered_measurements["volume_rescaled"] >= ninety_five_percentile]))
print("Number of OS in dataset smaller than 95 percentile: ", len(filtered_measurements[filtered_measurements["volume_rescaled"] < ninety_five_percentile]))
```

```python
filtered_measurements = filtered_measurements[filtered_measurements["volume_rescaled"] < ninety_five_percentile]
filtered_measurements
```

```python
np.max(filtered_measurements["volume_rescaled"])
```

```python
filtered_measurements.to_csv(path + "05-c-measurements_filtered-5-and-95-percentile.csv", index=False)
```
