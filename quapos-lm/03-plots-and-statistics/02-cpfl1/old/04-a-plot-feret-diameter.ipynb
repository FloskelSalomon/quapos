## Plot of the feret diameter
- let's investigate the average feret diameter in the cpfl1 model

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
path = "../../measurements/cpfl/"
save_path = "../../plots-images/quapos-paper/figure-3-plots/"
```

```python
measurements_averaged = pd.read_csv(path + "06-means-for-correlation-matrix.csv")
measurements_averaged = measurements_averaged.drop(labels=["label"], axis=1)
measurements_averaged
```

```python
feret_diameter = measurements_averaged[["age", "feret_diameter_rescaled", "genotype"]]
feret_diameter
```

```python
barplot = sns.barplot(data=feret_diameter,
                      x="age",
                      y="feret_diameter_rescaled",
                      hue="genotype",
                      palette={"wt": "m", "cpfl": "g"},
                      alpha=0.5,
                      hue_order=["wt", "cpfl"],
                      capsize=0.1,
                      errwidth=2.5,
                      width=0.8,
                      errorbar=("se"),
                      edgecolor="#000000",
                      linewidth=2.0)

stripplot = sns.stripplot(data=feret_diameter,
                          x="age",
                          y="feret_diameter_rescaled",
                          dodge=True,
                          hue="genotype",
                          palette={"wt": "m", "cpfl": "#025502"},
                          hue_order=["wt", "cpfl"])

barplot.legend_.remove()

plt.ylabel("Average OS length [$\mu$m]")
plt.ylim(0, 6)
plt.xlabel("Postnatal age in days")
plt.xticks(ticks=[0, 1, 2, 3, 4, 5], labels=[8, 14, 20, 30, 70, 245])

plt.savefig(save_path + "03-feret-diameter.png",
            dpi=300,
            bbox_inches="tight")
```
