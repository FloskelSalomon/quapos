## Filtered images

Supplementary figure 2 of the paper shows different filters which were used by the pixel classifier. Here these images for the figure are computed.

Here the following images are shown:

- intensity image
- normalised image
- gaussian blur (1) image
- difference of gaussian blur (1) image
- laplace box of gaussian blur (1) image

```python
import matplotlib.pyplot as plt
import numpy as np
import pyclesperanto_prototype as cle

import sys
sys.path.append("../../")
from quapos_lm import scale_bar_image, normalise_image
```

```python
# Define path to save the images
save_path = "../../plots-images/quapos-paper/sup-fig-2-images/"
```

```python
# Open the image 
# Same image as used for model validation, training data in figure 2
image = cle.imread("../../data/02-data-for-pixel-classifier/training-data/b-cropped-renamed/g.tif")
```

#### Normalised image

```python
# Normalise the image
normalised_image = normalise_image(image=image)

# Define figsize and subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 10))

# Remove unwanted subplots slot
axs[1].remove()

# Show intensity image and scale bar
cle.imshow(normalised_image[15, 10:90, 0:80], plot=axs[0])

# Remove ticks
axs[0].tick_params(left = False,
                right = False,
                labelleft = False,
                labelbottom = False, 
                bottom = False)

plt.savefig(save_path + "01-b-training-data-normalised.png",
            dpi=300,
            bbox_inches="tight")
```

#### Gaussian blur image

```python
# Compute gaussian blur
gaussian_blur = cle.gaussian_blur(source=normalised_image,
                                  sigma_x=1,
                                  sigma_y=1,
                                  sigma_z=1)

# Define figsize and subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 10))

# Remove unwanted subplots slot
axs[1].remove()

# Show intensity image and scale bar
cle.imshow(gaussian_blur[15, 10:90, 0:80], plot=axs[0])

# Remove ticks
axs[0].tick_params(left = False,
                right = False,
                labelleft = False,
                labelbottom = False, 
                bottom = False)

plt.savefig(save_path + "01-c-training-data-gaussian-blur.png",
            dpi=300,
            bbox_inches="tight")
```

#### Difference of gaussian blur

```python
# Compute differnce of gaussian blur image
dog = cle.difference_of_gaussian(source=normalised_image,
                                 sigma1_x=1,
                                 sigma1_y=1,
                                 sigma1_z=1,
                                 sigma2_x=2,
                                 sigma2_y=2,
                                 sigma2_z=2)

# Define figsize and subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 10))

# Remove unwanted subplots slot
axs[1].remove()

# Show intensity image and scale bar
cle.imshow(dog[15, 10:90, 0:80], plot=axs[0])

# Remove ticks
axs[0].tick_params(left = False,
                right = False,
                labelleft = False,
                labelbottom = False, 
                bottom = False)

plt.savefig(save_path + "01-d-training-data-dog.png",
            dpi=300,
            bbox_inches="tight")
```

#### Laplace box of gaussian blur

```python
# Compute laplace box of guassian blur
laplace_image = cle.laplace_box(source=normalised_image)
log = cle.gaussian_blur(source=laplace_image,
                        sigma_x=1,
                        sigma_y=1,
                        sigma_z=1)

# Define figsize and subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 10))

# Remove unwanted subplots slot
axs[1].remove()

# Show intensity image and scale bar
cle.imshow(log[15, 10:90, 0:80], plot=axs[0])

# Remove ticks
axs[0].tick_params(left = False,
                right = False,
                labelleft = False,
                labelbottom = False, 
                bottom = False)

plt.savefig(save_path + "01-d-training-data-log.png",
            dpi=300,
            bbox_inches="tight")
```
