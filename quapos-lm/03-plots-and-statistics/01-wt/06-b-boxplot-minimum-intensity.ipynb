## Boxplot of the minimum intensity

Additionally, it might be interesting to investigate the raw unaveraged data. Here, a boxplot of the minimum intensity will be computed were all datapoints are pooled.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Define path were measurements are stored
stats_path = "../../measurements/wt-postnatal-development/"

# Define path were plots should be stored
save_path = "../../plots-images/quapos-paper/sup-fig-3-plots/"
```

```python
# Load measurements
measurements = pd.read_csv(stats_path + "05-b-measurements-filtered.csv")
measurements
```

### Compute boxplot

```python
# Create boxplot
boxplot = sns.boxplot(data=measurements,
                            x="age",
                            y="minimum",
                            hue="age",
                            legend=False,
                            palette="deep")

# Adjust y axis
plt.yticks(fontsize=12)
plt.ylabel("OS minimum intensity", fontsize=14)

# Adjust x axis
plt.xticks(fontsize=12)
plt.xlabel("Postnatal age in days", fontsize=14)

plt.savefig(save_path + "03-b-minimum-boxplot.png",
            dpi=300,
            bbox_inches="tight")
```
