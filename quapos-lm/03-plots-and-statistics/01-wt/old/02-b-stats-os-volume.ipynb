## An anova test for the summed os volume throughout postnatal development is calculated

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.stats.multicomp as mc
```

```python
stats_path = "../measurements/wt-postnatal-development/"
```

```python
measurements = pd.read_csv(stats_path + "05-a-measurements-inf-reduced.csv")
measurements
```

```python
summed_volume = measurements.groupby("image_id", as_index=False)["volume_rescaled"].sum()
summed_volume
```

```python
age = measurements.groupby("image_id", as_index=False)["age", "biol_repl"].max()
age
```

```python
summed_volume = pd.merge(left=summed_volume,
                         right=age,
                         how="inner")
summed_volume
```

```python
volume_averaged = summed_volume.groupby("biol_repl", as_index=False)["volume_rescaled", "age"].mean()
volume_averaged
```
