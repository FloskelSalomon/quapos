## Prediction of postnatal development

The notebook aims to show a predicted image from the postnatal development of the ages P08, 12, 16, 20, and 24. 

The intensity images should be shown in the first row, corresponding predictions in the second row. The resulting images will be used in figure 2 of the paper.

Additionally, a function was written which computes an image with a scale bar, and can be overlayed on top of the actual image using `plt.subplots()`.

```python
import apoc
import matplotlib.pyplot as plt
import numpy as np
import pyclesperanto_prototype as cle

import sys
sys.path.append("../../")
from quapos_lm import scale_bar_image, predict_image
```

```python
# Load classifier
quapos_lm = apoc.ObjectSegmenter(opencl_filename = "../../01-training-and-validation/02-quapos-lm.cl")
quapos_lm.feature_importances()
```

The following representative images were used for figure 1 of the paper. Crops of these images will now be used to compute and visualise the prediction:

- `C1-p08-20x-zoom-3.5-replicate-3.tif`
- `C1-p12-20x-zoom-flo-2.6-replicate-9.tif`
- `C1-p16-20x-zoom-suse-1.5-replicate-19.tif`
- `C1-p20-20x-zoom-flo-1.4-replicate-21.tif`
- `C1-p24-20x-zoom-1.6-replicate-25.tif`

```python
# Define path where data is stored
images = "../../data/02-data-for-pixel-classifier/statistics-data/"

# Define path where images should be stored
save_path = "../../plots-images/quapos-paper/figure-2-images/"
```

### Predict masks from the images

Now the images will be loaded one after another, the masks predicted, and visualised. Addtionally, the first image will be used to define a scale bar. Since the microscope and objective for all images was the same the same scale bar can be applied to all images.

Only a crop of the image and prediction will be shown.

#### Image P08

```python
# Open image
image_p08 = cle.imread(images + "C1-p08-20x-zoom-3.5-replicate-3.tif")

# predict the image
prediction_p08 = predict_image(image=image_p08, classifier=quapos_lm)

# Compute scale bar of the crop
scale_bar, cmap = scale_bar_image(image=image_p08[15, 0:200, 600:800],
                                  micron_length=0.323,
                                  scale_length=25)
```

```python
# Define subplots and figure size
fig, axs = plt.subplots(2, 1, figsize=(5, 10))

# Show the intensity image
cle.imshow(image_p08[..., 0:200, 600:800], plot=axs[0])

# Add scale bar to intensity image
cle.imshow(scale_bar, plot=axs[0], colormap=cmap)

# Remove ticks and labels from the image
axs[0].tick_params(left=False,
                right=False,
                labelleft=False,
                labelbottom=False, 
                bottom=False)
    
# Show the corresponding prediction
cle.imshow(prediction_p08[..., 0:200, 600:800], plot = axs[1], labels = True)
axs[1].tick_params(left = False,
                right = False,
                labelleft = False,
                labelbottom = False, 
                bottom = False)

# Adjust the space between the images
plt.subplots_adjust(hspace=0.03)

plt.savefig(save_path + "01-a-prediction-p08.png",
            dpi=300,
            bbox_inches="tight")
```
