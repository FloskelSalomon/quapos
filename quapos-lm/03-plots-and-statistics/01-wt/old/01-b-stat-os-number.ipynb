## An anova test for the os number throughout postnatal development is calculated

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.stats.multicomp as mc
```

```python
stats_path = "../measurements/wt-postnatal-development/"
```

```python
measurements = pd.read_csv(stats_path + "05-c-measurements_filtered-5-and-95-percentile.csv")
measurements
```

```python
measurements_filtered = measurements[["label", "age", "biol_repl", "image_id"]]
measurements_filtered
```

```python
label = measurements_filtered.groupby("image_id", as_index=False)["label"].size()
label
```

```python
age = measurements.groupby("image_id", as_index=False)["age", "biol_repl"].max()
age
```

```python
grouped_dataframe = pd.merge(left=label,
                 right=age,
                 how="inner")
grouped_dataframe.rename(columns={"size":"label"}, inplace=True)
grouped_dataframe
```

```python
label_mean = grouped_dataframe.groupby("biol_repl", as_index=False)["label", "age"].mean()
label_mean
```

```python
#group by age and retrieve max labels
labels_grouped = label_mean.groupby("age")["label"]

#weird for loop that itterates them all into seperate arrays
groups = [labels_grouped.get_group(group).values for group in labels_grouped.groups]

#show the arrays
groups
```

```python
#calculate f value and p value from scipy stats function
f_value, p_value = stats.f_oneway(*groups)

#print the values
print("F-value:", f_value)
print("p-value:", p_value)
```

```python
#posthoc results
posthoc_results = mc.MultiComparison(
    label_mean["label"], 
    label_mean["age"]
)

#tukey test
tukey_results = posthoc_results.tukeyhsd()

tukey_results.summary()
```
