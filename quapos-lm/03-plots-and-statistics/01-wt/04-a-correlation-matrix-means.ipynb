## Compute Correlation Matrix

To investigate the relationships of all measured features a correlation matrix will be computed. The computation is based on functions from `pandas` and `seaborn`.

The correlation matrix will be calculated with the values averaged by `image_id`, and `biol_repl`.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

import sys
sys.path.append("../../")
from quapos_lm import correlation_matrix
```

```python
# Define path where measurements are stored
path = "../../measurements/wt-postnatal-development/"

# Define path to save plots
save_path = "../../plots-images/quapos-paper/sup-fig-3-plots/"
```

```python
# Load measurements
measurements = pd.read_csv(path + "09-renamed-for-correlation.csv")

# Move column age to the end of the dataframe
# Should be displayed as bottom row in correlation matrix
cols = list(measurements.columns)
cols.append(cols.pop(cols.index("age")))
measurements = measurements[cols]
measurements
```

### Compute the correlation matrix

The correlation matrix is now computed with a predefined function. The colorbar will be removed here (note: in the supplementary figure 3 of the paper the colorbar will be shown next to the correlation vector of the age).

Below you find the same correlation matrix with the colorbar.

```python
# Define figure size
plt.figure(figsize=(8, 7))

# Compute heatmap representation of correlation matrix
heatmap = correlation_matrix(measurements, font_scale=1)

# Remove colorbar
heatmap.collections[0].colorbar.remove()

plt.savefig(save_path + "01-a-correaltion-matrix.png",
            dpi=300,
            bbox_inches="tight")
```
