## Boxplot of the BBox height

Additionally, it might be interesting to investigate the raw unaveraged data. Here, a boxplot of the bounding box height will be computed were all datapoints are pooled.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Define path where measurements are stored
stats_path = "../../measurements/wt-postnatal-development/"

# Define path were plots should be stored
save_path = "../../plots-images/quapos-paper/figure-2-plots/"
```

```python
# Load measurements
measurements = pd.read_csv(stats_path + "05-b-measurements-filtered.csv")
measurements
```

### Create the boxplot

```python
# Create boxplot
boxplot = sns.boxplot(data=measurements,
                            x="age",
                            y="bbox_4_rescaled",
                            hue="age",
                            legend=False,
                            palette="deep")

# Adjust y axis
plt.yticks(fontsize=12)
plt.ylabel("OS bbox height [$\mu$m]", fontsize=14)

# Adjust x axis
plt.xticks(fontsize=12)
plt.xlabel("Postnatal age in days", fontsize=14)

plt.savefig(save_path + "04-b-bbox-height-bp.png",
            dpi=300,
            bbox_inches="tight")
```
