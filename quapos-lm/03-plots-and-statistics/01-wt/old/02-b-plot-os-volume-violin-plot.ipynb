## Violin plot of the os volume

- a violin plot of the os volume will be computed
- filtered dataset (5-/95-percentiles)
- no averages beforehand

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
stats_path = "../../measurements/wt-postnatal-development/"
save_path = "../../plots-images/quapos-paper/sup-figure-3-plots/"
```

```python
measurements = pd.read_csv(stats_path + "05-b-measurements-filtered.csv")
measurements
```

```python
# Create violinplot
violinplot = sns.violinplot(data=measurements,
                            x="age",
                            y="volume_rescaled",
                            hue="age",
                            legend=False,
                            palette="deep")

# Adjust y axis
#plt.ylim(-1)
plt.yticks(fontsize=12)
plt.ylabel("OS volume [$\mu$m$^3$]", fontsize=14)

# Adjust x axis
plt.xticks(fontsize=12)
plt.xlabel("Postnatal age in days", fontsize=14)

#plt.savefig(save_path + "02-b-os-volume-violinplot.png",
#            dpi=300,
#            bbox_inches="tight")
```

```python
# Create boxplot
boxplot = sns.violinplot(data=measurements,
                            x="age",
                            y="feret_diameter_rescaled",
                            hue="age",
                            legend=False,
                            palette="deep")

# Adjust y axis
plt.ylim(-1)
plt.yticks(fontsize=12)
plt.ylabel("OS volume [$\mu$m$^3$]", fontsize=14)

# Adjust x axis
plt.xticks(fontsize=12)
plt.xlabel("Postnatal age in days", fontsize=14)

#plt.savefig(save_path + "02-c-os-volume-boxplot.png",
#            dpi=300,
#            bbox_inches="tight")
```
