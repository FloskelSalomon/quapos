## Summed OS volume
- summed OS volume without filtering will be investigated

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
path = "../../measurements/cpfl/"
save_path = "../../plots-images/quapos-paper/figure-3-plots/"
```

```python
measurements = pd.read_csv(path + "04-measurements-rescaled.csv")
measurements
```

### Preparing the dataset
- retrieving summed volume per image
- retrieve age in a df
- combine both dfs

```python
volume_sum = measurements.groupby("image_id", as_index=False)["volume_rescaled"].sum()
volume_sum
```

```python
age = measurements.groupby("image_id", as_index=False)["age", "genotype"].max()
age
```

```python
volume_sum = pd.merge(volume_sum,
                      age,
                      how="inner")
volume_sum
```

## Plot the volume

```python
barplot = sns.barplot(data=volume_sum,
                      x="age",
                      y="volume_rescaled",
                      hue="genotype",
                      palette={"wt": "m", "cpfl": "g"},
                      alpha=0.5,
                      hue_order=["wt", "cpfl"],
                      capsize=0.1,
                      errwidth=2.5,
                      width=0.8,
                      errorbar=("se"),
                      edgecolor="#000000",
                      linewidth=2.0)

stripplot = sns.stripplot(data=volume_sum,
                          x="age",
                          y="volume_rescaled",
                          dodge=True,
                          hue="genotype",
                          palette={"wt": "m", "cpfl": "#025502"},
                          hue_order=["wt", "cpfl"])

barplot.legend_.remove()

plt.ylim(0, 8000)

plt.ylabel("Average summed OS volume [$\mu$m$^3$]")
plt.xlabel("Postnatal age in days")

plt.savefig(save_path + "02-summed-os-volume-comparison.png",
            dpi=300,
            bbox_inches="tight")
```
